## Lab Work

### BGR, RGB, and Grayscale

#### Displaying the Difference Between Channel Orders

```python
cv2_demo_image = cv.imread(os.path.join(
    '..', data_dir_name, 'demo_RGB_image.png'))
plt.imshow(cv2_demo_image)

plt.show()
```

#### Swapping Color Represntation 

```python
demo_image_rgb = cv.cvtColor(cv2_demo_image, cv.COLOR_BGR2RGB)
plt.imshow(demo_image_rgb)

plt.show()
```

#### Displaying Grayscale Image

```python
demo_image_gray = cv.cvtColor(cv2_demo_image, cv.COLOR_BGR2GRAY)

fig, axs = plt.subplots(ncols=3, figsize=(12, 8))

axs[0].imshow(demo_image_rgb)
axs[0].set_title('input color image')
axs[0].set_axis_off()

axs[1].imshow(demo_image_gray)
axs[1].set_title('default view image')
axs[1].set_axis_off()

axs[2].imshow(demo_image_gray, cmap='gray')
axs[2].set_title('gray cmap')
axs[2].set_axis_off()

plt.show()
```

## Final Work

```python
messi = cv.imread(os.path.join('..', data_dir_name, 'messi.jpg'))
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_axis_off()
ax.imshow(cv.cvtColor(messi, cv.COLOR_BGR2RGB))

plt.show()
```

```python
# box around Messi
cv.rectangle(messi,
             (60,   50),     # top left
             (460, 340),     # bottom right
             (0, 0, 255))    # red

cv.putText(messi, 'Messi', (60, 45),
           cv.FONT_HERSHEY_COMPLEX_SMALL, 0.8, (255, 255, 255))

# box around the ball
cv.rectangle(messi,
             (330, 285),      # top left
             (390, 338),      # bottom right
             (255, 0, 0))     # blue

cv.putText(messi, 'ball', (330, 280),
           cv.FONT_HERSHEY_COMPLEX_SMALL, 0.8, (255, 255, 255))


messi = np.pad(messi, ((20, 0), (0, 0), (0, 0)),
               'constant', constant_values=255)

text = "Messi's in red, the ball is in blue, that leg twist looks painful, this pun attept, too"

cv.putText(messi, text, (5, 15), cv.FONT_HERSHEY_COMPLEX_SMALL, 0.5, 0)

fig, ax = plt.subplots(figsize=(20, 10))
ax.set_axis_off()
ax.imshow(cv.cvtColor(messi, cv.COLOR_BGR2RGB))
plt.show()
```
